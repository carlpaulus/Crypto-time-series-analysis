```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow
from sklearn.preprocessing import MinMaxScaler
```

```python
np.version.version
```

```python
PORTFOLIO = 'BTC-USD'
start = '2015-01-01'
end = '2021-09-1'
```

```python
dataset = pd.DataFrame(yf.download(PORTFOLIO, start=start, end=end).ffill().dropna())
```

```python
dataset
```

```python
dataset['Date'] = dataset.index
```

```python
dataset.reset_index(drop=True, inplace=True)
```

```python
dataset
```

```python
data_training = dataset[dataset['Date']< '2021-04-01'].copy()
```

```python
data_training
```

```python
data_test = dataset[dataset['Date'] > '2021-03-31'].copy()
```

```python
data_test
```

```python
training_data = data_training.drop(['Date', 'Adj Close'], axis = 1)
training_data.head()
```

```python
scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
X_train = [] 
Y_train = []
```

```python
training_data.shape
```

```python
for i in range(60, training_data.shape[0]):
    X_train.append(training_data[i-60:i])
    Y_train.append(training_data[i,0])
   
        
```

```python
X_train, Y_train = np.array(X_train), np.array(Y_train)
X_train.shape
```

```python
Y_train
```

```python
from tensorflow.keras import Sequential
```

```python
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
model = Sequential() 

```

```python
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1], 5)))
model.add(Dropout(0.2)) 
model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))
model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(units =1))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
history= model.fit(X_train, Y_train, epochs = 25, batch_size = 64, validation_split=0.1)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(loss))
plt.figure()
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title("Training and Validation Loss")
plt.legend()
plt.show()
```

```python
part_60_days = data_training.tail(60)
df= part_60_days.append(data_test, ignore_index = True)
df = df.drop(['Date', 'Adj Close'], axis = 1)
df.head()
```

```python
inputs = scaler.transform(df)
inputs
```

```python
X_test = []
Y_test = []
```

```python
for i in range (60, inputs.shape[0]):
    X_test.append(inputs[i-60:i])  
    Y_test.append(inputs[i, 0])

```

```python
X_test, Y_test = np.array(X_test), np.array(Y_test)
```

```python
Y_pred = model.predict(X_test)
```

```python
scaler.scale_
```

```python
scale = 1/1.63815631e-05
Y_test = Y_test*scale 
Y_pred = Y_pred*scale
Y_pred
```
